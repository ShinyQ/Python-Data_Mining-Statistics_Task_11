````
import math
import pandas as pd
import seaborn as sns
import numpy as np
````

### **Mencari Korelasi**

````
D1 = [1.8, 8.5, 8.4, 4.5, 4.3, 6.9, 8.0, 5.0, 3.6, 2.6, 2.6, 4.2, 3.6, 3.7]
D2 = [3.1, 0.8, 0.9, 1.5, 1.2, 1.0, 1.1, 1.3, 1.8, 1.9, 1.5, 1.4, 1.8, 2.1]
n = len(D1)
````

### **Data Modelling**

````
data = pd.DataFrame({'Peningkatan Upah': D1, 'Pengangguran':D2})
correlation = round(np.corrcoef(D1, D2)[1][0], 3)

total = 0
i = 0
for n in D1: 
     total = total + (n * D2[i])
     i = i + 1 

s3 = 0
for n in D1:
    s3 = s3 + (n * n)

s4 = 0
for n in D2:
    s4 = s4 + (n * n)

s1 = round(sum(D1), 2)
s2 = round(sum(D2), 2)
````

### **Hasil Data Korelasi**

````
print('ΣD1D2 =', total)
print('ΣD1 =', s1)
print('ΣD2 =', s2)
print('ΣD1^2 =', round(s3, 2))
print('ΣD2^2 =', s4)
print('Korelasi Dua Data :', correlation)
````

### **Scatterplot Data**

````
sns.scatterplot(data=data, x="Peningkatan Upah", y="Pengangguran")
````

### **Metode Pengujian Korelasi**

````
alpha = 0.05            # Signifikansi
jumlah = 14             # Jumlah Objek
koefisien = correlation

t = round(koefisien/(math.sqrt(((1-(koefisien*koefisien))/ (jumlah - 2)))), 3)
print('Nilai t =', t)
print('Nilai df =', jumlah-2)
print('α / 2 =', alpha/2)
````
